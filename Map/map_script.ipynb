## In this notebook, I will write a script to do some cleaning on the 'State-Territory-Dementia-2011-2020.xlsx' dataset to get the dataset required to create an interactive map on Tableau which will show the Statistics of Dementia Patients in Australia from year 2011-2020

First,let's import the necessary libraries

```python
import pandas as pd                                 # To manipulate dataframes
import warnings                                     # To ignore the warnings in order to beautify the script
warnings.filterwarnings('ignore')
```

### Now, let's read the 'State-Territory-Dementia-2011-2020.xlsx' and store it as df 

```python
df = pd.read_excel('State-Territory-Dementia-2011-2020.xlsx')
```

### Let's have a look at the dataset

```python
df
```

### So, the dataset has information of number of male, female and overall Dementia patients in Australia from year 2011 to 2020 in all the 8 States

### The next step is to change the name of the States from Abbreviation to the full form in order to match it with Shape file of the States

```python
for index,value in enumerate(df.States):
    if value == "NSW":
        new_value = 'New South Wales'
        df.States[index]=new_value
    if value == "Vic":
        new_value = 'Victoria'
        df.States[index]=new_value
    if value == "Qld":
        new_value = 'Queensland'
        df.States[index]=new_value
    if value == "WA":
        new_value = 'Western Australia'
        df.States[index]=new_value
    if value == "SA":
        new_value = 'South Australia'
        df.States[index]=new_value
    if value == "Tas":
        new_value = 'Tasmania'
        df.States[index]=new_value
    if value == "ACT":
        new_value = 'Australian Capital Territory'
        df.States[index]=new_value
    if value == "NT":
        new_value = 'Northern Territory'
        df.States[index]=new_value
```

```python
df.head()
```

### Now, the dataset looks good is ready to create the map in the Tableau. So, let's export the dataset 

```python
df.to_excel('stats.xlsx')
```
